```python
import random

# Generate random number from 1 to 20
secret_number = random.randint(1, 20)

print("I'm thinking of a number between 1 and 20.")
print("You have 5 attempts to guess it!")

# Use a for loop to give limited attempts
for attempt in range(1, 6):
    guess = int(input(f"Attempt {attempt}: Enter your guess: "))

    if guess < secret_number:
        print("Too low.")
    elif guess > secret_number:
        print("Too high.")
    else:
        print("Correct! You guessed the number!")
        break  # Stop the loop if the player is correct

# If the loop finishes without a break, the player failed to guess
else:
    print(f"Sorry, you are out of attempts. The secret number was {secret_number}.")

```
