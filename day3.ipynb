```python
# Example 4: Conditions with logical operators (and, or, not)
has_ticket = False
has_id = False
print(f"Has Ticket: {has_ticket}, Has ID: {has_id}")
if has_ticket and has_id:
    print("entry allowed")
elif has_ticket or has_ticket:
    print("partial requirements is misssing")
    if not has_id:
        print("warning:id is missing")
else:
    print("entry denied")

    
```
